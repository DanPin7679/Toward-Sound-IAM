```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from scipy.integrate import solve_ivp
```

```python
class System:
    def __init__(self, name, t_eval, stocks):
        self.name = name
        self.t_eval = t_eval
        self.stocks = stocks
        
    def run(self):
        qts = []
        
        for s in self.stocks:
            qts.append(s.qt)
        
        def dx_dt(t, x): 
            dotx = self.stocks[0].dx_dt(t, x[0])
            doty = self.stocks[1].dx_dt(t, x[1])
            return np.array([dotx, doty])
            
        sol = solve_ivp(dx_dt, [0, 131], qts, t_eval=t_eval)
        return sol
```

```python
class Stock:
    def __init__(self, name, qt, flow_in, flow_out, is_loop=True):
        self.name = name
        self.qt = qt
        self.flow_in = flow_in
        self.flow_out = flow_out
        self.is_loop = is_loop
        
    def dx_dt(self, t, x):
        f_in = self.flow_in.calculate()
        f_out = self.flow_out.calculate()
        return (f_in - f_out) * (x if self.is_loop else 1)
```

```python
class Flow:
    def __init__(self, name, qt):
        self.name = name
        self.qt = qt
        
    def calculate(self):
        return self.qt
```

```python
def drawStockEvol(t, xs, stock):
    plt.figure(figsize = (12, 4))
    for x in xs:
        plt.plot(t, x)
    plt.xlabel('Year')
    plt.ylabel(stock)
    plt.show()
    
```

```python
p_init = 1.65e9
BR = Flow("Birth Rate", 0.04)
DR = Flow("Death Rate", 0.02)
pop = Stock("Population", p_init, BR, DR, True)
```

```python
p_init2 = 0.5e9
BR2 = Flow("Birth Rate", 0.05)
DR2 = Flow("Death Rate", 0.02)
pop2 = Stock("Population", p_init2, BR2, DR2, True)
```

```python
t_eval = np.arange(0, 130, 1)
syst = System("World2", t_eval, [pop, pop2])
```

```python
sol2 = syst.run()
drawStockEvol(sol2.t + 1970, [sol2.y[0], sol2.y[1]], 'Population')
```
