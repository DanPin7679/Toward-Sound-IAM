```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from scipy.integrate import solve_ivp
```

```python
class Stock:
    def __init__(self, name, qt, flow_in, flow_out):
        self.name = name
        self.qt = qt
        self.flow_in = flow_in
        self.flow_out = flow_out
        
    def dx_dt(self):
        f_in = self.flow_in.calculate()
        f_out = self.flow_out.calculate()
        return lambda t,x: (f_in - f_out) * x
```

```python
class Flow:
    def __init__(self, name, qt):
        self.name = name
        self.qt = qt
        
    def calculate(self):
        return self.qt
```

```python
def drawStockEvol(t, xs, stock):
    plt.figure(figsize = (12, 4))
    for x in xs:
        plt.plot(t, x)
    plt.xlabel('Year')
    plt.ylabel(stock)
    plt.show()
    
t_eval = np.arange(0, 130, 1)
```

```python
p_init = 1.65e9
BR = Flow("Birth Rate", 0.04)
DR = Flow("Death Rate", 0.02)
pop = Stock("Population", p_init, BR, DR)
```

```python
sol = solve_ivp(pop.dx_dt(), [0, 131], [pop.qt], t_eval=t_eval)
drawStockEvol(sol.t + 1970, [sol.y[0]], 'Population')
```
